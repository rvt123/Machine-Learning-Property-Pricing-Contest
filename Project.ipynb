```python
with open('assessor0.0.json') as json_data:
    data = json.load(json_data)
```

```python
raw_dataframe=pd.DataFrame({ key:pd.Series(value) for key, value in data.items() })
```

```python
raw_dataframe.head()
```

```python
raw_dataframe.columns
```

```python
reObj = re.compile('name')
column_name=[]
for count in range(0,raw_dataframe.fields.notna().sum()):
    for key in (raw_dataframe.fields[count]).keys():
        if(reObj.match(key)):
            column_name.append((raw_dataframe.fields[count])[key])        
```

```python
column_name
```

```python
dataframe=pd.DataFrame(columns=column_name)
```

```python
for colname in column_name:
    reObj_row = re.compile(r'\b%s\b'%colname)
    coldata=[]
    for count in range(0,len(raw_dataframe.features)):
        for key in (raw_dataframe.features[count]['attributes']).keys():
            if(reObj_row.match(key)):
                coldata.append((raw_dataframe.features[count]['attributes'])[key])
    dataframe[colname]=coldata    
```

```python
with open('assessor79.0.json') as json_data:
    data = json.load(json_data)
raw_dataframe=pd.DataFrame({ key:pd.Series(value) for key, value in data.items() })
raw_dataframe.head()
```

```python
reObj = re.compile('name')
column_name_1=[]
for count in range(0,raw_dataframe.fields.notna().sum()):
    for key in (raw_dataframe.fields[count]).keys():
        if(reObj.match(key)):
            column_name_1.append((raw_dataframe.fields[count])[key]) 
            
column_name_1
```

```python
dataframe_1=pd.DataFrame(columns=column_name_1)
```

```python
for colname in column_name_1:
    reObj_row = re.compile(r'\b%s\b'%colname)
    coldata=[]
    for count in range(0,len(raw_dataframe.features)):
        for key in (raw_dataframe.features[count]['attributes']).keys():
            if(reObj_row.match(key)):
                coldata.append((raw_dataframe.features[count]['attributes'])[key])
    dataframe_1[colname]=coldata    
```

```python
dataframe=pd.concat([dataframe,dataframe_1],join='inner')
dataframe.reset_index(inplace=True)
```
